```python
from datasets import load_dataset
dataset = load_dataset("Joanne/cardillo_2010_2017_random")
```

```python
Train_data=[]
Test_data=[]
for i in dataset["train"]:
  temp={}
  temp["instruction"] = "Given the statement find A and B values, where A dictates what we are talking about and B tells what about it, also return in structured output"
  temp["input"] = i["sentence"]
  temp["output"] = "{A : "+ i["A"]+", "+"B: "+i["B"]+"}"
  Train_data.append(temp)

for i in dataset["test"]:
  temp={}
  temp["instruction"] = "Given the statement find A and B values, where A dictates what we are talking about and B tells what about it, also return in structured output"
  temp["input"] = i["sentence"]
  temp["output"] = "{A : "+ i["A"]+", "+"B: "+i["B"]+"}"
  Test_data.append(temp)
```

```python
import json
json_string = json.dumps(Train_data)
with open('Train_data.json', 'w') as file:
    json.dump(Train_data, file)
```

```python
import pandas as pd
df= pd.DataFrame(Train_data)
df.to_csv("Training_csv.csv",index=False)

df= pd.DataFrame(Test_data)
df.to_csv("Test_csv.csv",index=False)
```
